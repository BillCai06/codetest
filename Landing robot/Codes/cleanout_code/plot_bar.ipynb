```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('MCTS_6p_10000.csv')

# Calculate average and standard deviation
median_duration = df['duration_mcts_v2'].median()
avg_duration = df['duration_mcts_v2'].mean()
stdv_duration = df['duration_mcts_v2'].std()
min_duration = df['duration_mcts_v2'].min()
max_duration = df['duration_mcts_v2'].max()

# print(median_duration, avg_duration) # they're about the same

median_total_time = df['total_time'].median()
avg_total_time = df['total_time'].mean()
stdv_total_time = df['total_time'].std()
min_total_time = df['total_time'].min()
max_total_time = df['total_time'].max()
```

```python
# Set up the subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 10))

# Increase the width of the fill
offset = 0.2

# Plotting for duration_mcts_v2
x_vals = [-offset, offset]
axs[0].fill_between(x_vals, [median_duration-stdv_duration]*2, [median_duration+stdv_duration]*2, color='lightgray', label='Standard Deviation')
axs[0].errorbar(0, median_duration, yerr=[[median_duration-min_duration], [max_duration-median_duration]], fmt='o', color='blue', ecolor='black', capsize=10, label='Median with Min-Max')
# axs[0].errorbar(0, avg_duration, yerr=stdv_duration, fmt='o', color='blue', ecolor='black', capsize=10, label='Mean with Error')
axs[0].set_title('MCTS Runtime 10000 iterations (100 Datapoints)')
axs[0].set_xticks([0])
axs[0].set_xticklabels(['MCTS Runtime'])
axs[0].set_ylabel('Seconds')
axs[0].grid(True, which='both', axis='y')
axs[0].legend()

# Plotting for total_time
x_vals = [-offset, offset]
axs[1].fill_between(x_vals, [median_total_time-stdv_total_time]*2, [median_total_time+stdv_total_time]*2, color='lightgray', label='Standard Deviation')
axs[1].errorbar(0, median_total_time, yerr=[[median_total_time-min_total_time], [max_total_time-median_total_time]], fmt='o', color='blue', ecolor='black', capsize=10, label='Median with Min-Max')
# axs[1].errorbar(0, avg_total_time, yerr=stdv_total_time, fmt='o', color='blue', ecolor='black', capsize=10, label='Mean with Error')
axs[1].set_title('Mission Time')
axs[1].set_xticks([0])
axs[1].set_xticklabels(['Mission Time'])
axs[1].set_ylabel('Hours')
axs[1].grid(True, which='both', axis='y')
axs[1].legend()

# Display the plots
plt.tight_layout()
plt.show()
```

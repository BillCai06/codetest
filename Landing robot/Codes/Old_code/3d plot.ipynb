```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define rotation matrices
rot_matrices = [
    np.array([[ 0.9966,  0.079 ,  0.0206],
              [-0.0794,  0.9966,  0.019 ],
              [-0.019 , -0.0206,  0.9996]]),
    np.array([[ 0.995 , -0.0786, -0.0602],
              [ 0.0798,  0.9966,  0.0174],
              [ 0.0586, -0.0222,  0.998 ]]),
    np.array([[ 9.998e-01, -4.000e-04,  1.980e-02],
              [-4.000e-04,  9.992e-01,  3.960e-02],
              [-1.980e-02, -3.960e-02,  9.990e-01]])
]

# Set up the figure and axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the origin
ax.scatter(0, 0, 0, c='k', marker='o', s=50)

# Plot each rotation matrix as a set of three vectors (columns of the matrix)
colors = ['r', 'g', 'b']
labels = ['X', 'Y', 'Z']

for idx, rot_matrix in enumerate(rot_matrices):
    for j, column in enumerate(rot_matrix.T):
        ax.quiver(0, 0, 0, column[0], column[1], column[2], color=colors[j], length=1, linewidth=1.5, arrow_length_ratio=0.1)

# Set the aspect ratio and the limits
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.5, 1.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("3D Visualization of Rotation Matrices")
ax.view_init(elev=30, azim=30)

# Show the plot
plt.legend()
plt.show()
```

```python
from IPython.display import clear_output
from matplotlib import pyplot as plt
import numpy as np
import collections
%matplotlib inline

def live_plot(data_dict, figsize=(7,5), title=''):
    clear_output(wait=True)
    plt.figure(figsize=figsize)
    for label,data in data_dict.items():
        plt.plot(data, label=label)
    plt.title(title)
    plt.grid(True)
    plt.xlabel('epoch')
    plt.legend(loc='center left') # the plot evolves to the right
    plt.show();
```

```python
# Interactive plot for visualizing rotation matrices sequentially

from IPython.display import display, clear_output

def plot_rotation(rot_matrix, ax):
    """Plot a rotation matrix."""
    # Clear the previous plot
    ax.clear()
    
    # Plot the origin
    ax.scatter(0, 0, 0, c='k', marker='o', s=50)

    # Plot the rotation matrix as a set of three vectors (columns of the matrix)
    for j, column in enumerate(rot_matrix.T):
        ax.quiver(0, 0, 0, column[0], column[1], column[2], color=colors[j], length=1, label=f'{labels[j]}', linewidth=1.5, arrow_length_ratio=0.1)

    # Set the aspect ratio and the limits
    ax.set_xlim([-1.5, 1.5])
    ax.set_ylim([-1.5, 1.5])
    ax.set_zlim([-1.5, 1.5])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title("3D Visualization of Rotation Matrix")
    ax.view_init(elev=30, azim=30)
    plt.legend()

# Set up the figure and axis
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Display each rotation matrix with a brief pause in between
for rot_matrix in rot_matrices:
    plot_rotation(rot_matrix, ax)
    plt.pause(1.0)
    clear_output(wait=True)
    display(fig)
    plt.pause(1.0)

plt.close()
```
